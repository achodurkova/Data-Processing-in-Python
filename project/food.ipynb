```python
import pandas as pd

df = pd.DataFrame(rows)
```

```python
import requests
import pandas as pd
from datetime import datetime, timedelta
import time

base_url = "https://kamweb.ruk.cuni.cz/webkredit/Api/Ordering/Menu"

canteen_id = 7
start = datetime(2026, 4, 14)

all_rows = []

for i in range(7):
    day = start + timedelta(days=i)
    date_str = day.strftime("%Y-%m-%dT22:00:00.000Z")

    r = requests.get(base_url, params={
        "Dates": date_str,
        "CanteenId": canteen_id
    })

    data = r.json()

    for group in data["groups"]:
        if not group.get("rows"):
            continue  # skip empty groups
        for row in group["rows"]:
            item = row["item"]

            all_rows.append({
                "date": group["date"],
                "meal": item["mealName"],
                "price": item["price2"],
                "type": item["mealKindName"]
            })
        time.sleep(1)

df = pd.DataFrame(all_rows)
```

```python
df
```

```python
df.to_csv("menza_week.csv", index=False)
```
