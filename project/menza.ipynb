```python
import numpy as np
import pandas as pd 
import os
import requests
import json 
from bs4 import BeautifulSoup
import boto3
from IPython.display import HTML, display
import time 
import datetime 
```

```python
#create a list of days in a week
days=list(range(19,24))
data=[]
for day in days:
    url=f"https://kamweb.ruk.cuni.cz/webkredit/Api/Ordering/Menu?Dates=2026-04-{day}T22%3A00%3A00.000Z&CanteenId=7"
    response=requests.get(url).json()

    #loop through each meal type group to get type and date
    for group in response["groups"]:
        meal_kind=group["mealKindName"]
        date=group["date"]
        
        #loop though each option to get details
        for row in group["rows"]:
            item=row["item"]
            data.append({
                'date': date,
                'meal_type': meal_kind,
                'name': item['mealName'],
                'price': item['price2']
            })
    time.sleep(1)
```

```python
df=pd.DataFrame(data)
df
```

```python
df["date"]=df["date"].str.split("T").str[0]
df=df[df["name"]!="Teplý bufet"]
df=df.reset_index(drop=True)
df
```

```python
df.to_csv("menza_prices.csv",index=False)
```
